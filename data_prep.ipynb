```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/treningowy anti-AR.xlsx")
```

```python
df
```

```python
df['TPC']
```

```python
if "AMW" in df.columns and "TPC" in df.columns:
    # obliczenie korelacji Pearsona
    corr = df["AMW"].corr(df["TPC"])
    print(f"Korelacja Pearsona między AMW a TPC: {corr:.4f}")
else:
    print("W DataFrame brakuje kolumn 'AMW' lub 'TPC'")
```

```python
# Spearman
corr_spearman = df["AMW"].corr(df["TPC"], method="spearman")

# Kendall
corr_kendall = df["AMW"].corr(df["TPC"], method="kendall")
```
